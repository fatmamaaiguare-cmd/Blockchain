```python
import hashlib
import time
from datetime import datetime

# =============================================================================
# Ã‰TAPE 1: DÃ‰FINITION DE LA CLASSE TRANSACTION
# =============================================================================
class Transaction:
    """
    ReprÃ©sente une transaction individuelle dans la blockchain
    """
    def __init__(self, from_addr, to_addr, amount):
        self.from_addr = from_addr  # Adresse expÃ©diteur (None pour rÃ©compenses)
        self.to_addr = to_addr      # Adresse destinataire
        self.amount = amount        # Montant de la transaction

    def __str__(self):
        return f"Transaction({self.from_addr} â†’ {self.to_addr}: {self.amount})"

# =============================================================================
# Ã‰TAPE 2: DÃ‰FINITION DE LA CLASSE BLOCK
# =============================================================================
class Block:
    """
    ReprÃ©sente un bloc dans la blockchain
    """
    def __init__(self, timestamp, transactions, previous_hash=''):
        self.timestamp = timestamp          # Horodatage de crÃ©ation
        self.transactions = transactions    # Liste des transactions
        self.previous_hash = previous_hash  # Hash du bloc prÃ©cÃ©dent (lien)
        self.nonce = 0                      # Nombre utilisÃ© pour le minage
        self.hash = self.create_hash()      # Hash actuel du bloc

    def create_hash(self):
        """
        CrÃ©e le hash cryptographique du bloc
        Utilise SHA-256 pour gÃ©nÃ©rer une empreinte unique
        """
        # SÃ©rialisation de toutes les donnÃ©es du bloc
        transactions_str = str([t.__dict__ for t in self.transactions])
        block_string = (str(self.previous_hash) +
                       str(self.timestamp) +
                       transactions_str +
                       str(self.nonce)).encode()

        return hashlib.sha256(block_string).hexdigest()

    def mine_block(self, difficulty):
        """
        Processus de minage (Preuve de Travail)
        Trouve un nonce tel que le hash commence par 'difficulty' zÃ©ros
        """
        target = '0' * difficulty  # Cible Ã  atteindre (ex: "00" pour difficultÃ©=2)

        print(f"â›ï¸  Minage en cours... Recherche hash commenÃ§ant par '{target}'")
        start_time = time.time()

        while self.hash[:difficulty] != target:
            self.nonce += 1
            self.hash = self.create_hash()

        mining_time = time.time() - start_time
        print(f"âœ… Bloc minÃ© ! Nonce={self.nonce}, Temps={mining_time:.2f}s")
        print(f"ğŸ”— Hash final: {self.hash}")

# =============================================================================
# Ã‰TAPE 3: DÃ‰FINITION DE LA CLASSE BLOCKCHAIN
# =============================================================================
class Blockchain:
    """
    Classe principale gÃ©rant toute la blockchain
    """
    def __init__(self):
        self.chain = [self.create_genesis_block()]  # ChaÃ®ne de blocs
        self.difficulty = 2                         # DifficultÃ© de minage
        self.pending_transactions = []              # Transactions non confirmÃ©es
        self.mining_reward = 10                     # RÃ©compense du mineur

    def create_genesis_block(self):
        """
        CrÃ©e le premier bloc de la blockchain (bloc genÃ¨se)
        """
        print("ğŸ¯ CrÃ©ation du bloc genÃ¨se...")
        genesis_block = Block(
            timestamp="01/01/2020 00:00:00",
            transactions=[],
            previous_hash="0"
        )
        print(f"âœ… Bloc genÃ¨se crÃ©Ã© - Hash: {genesis_block.hash}")
        return genesis_block

    def get_last_block(self):
        """
        Retourne le dernier bloc de la chaÃ®ne
        """
        return self.chain[-1]

    def create_transaction(self, transaction):
        """
        Ajoute une nouvelle transaction Ã  la file d'attente
        """
        print(f"ğŸ“ Ajout transaction: {transaction}")
        self.pending_transactions.append(transaction)

    def mine_pending_transactions(self, miner_address):
        """
        Minage des transactions en attente et crÃ©ation d'un nouveau bloc
        """
        print(f"\nğŸ¯ DÃ‰BUT MINAGE - Mineur: {miner_address}")
        print(f"ğŸ“‹ Transactions Ã  miner: {len(self.pending_transactions)}")

        # Ã‰tape 1: CrÃ©ation du nouveau bloc
        current_time = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
        block = Block(
            timestamp=current_time,
            transactions=self.pending_transactions[:],  # Copie des transactions
            previous_hash=self.get_last_block().hash
        )

        # Ã‰tape 2: Minage du bloc
        block.mine_block(self.difficulty)

        # Ã‰tape 3: Ajout Ã  la chaÃ®ne
        self.chain.append(block)
        print(f"ğŸ“¦ Bloc #{len(self.chain)-1} ajoutÃ© Ã  la chaÃ®ne")

        # Ã‰tape 4: RÃ©compense du mineur
        reward_transaction = Transaction(None, miner_address, self.mining_reward)
        self.pending_transactions = [reward_transaction]

        print(f"ğŸ’° RÃ©compense: {self.mining_reward} unitÃ©s â†’ {miner_address}")
        print(f"ğŸ“‹ Nouvelles transactions en attente: {len(self.pending_transactions)}")

    def get_balance_of_address(self, address):
        """
        Calcule le solde d'une adresse en parcourant toute la blockchain
        """
        balance = 0
        print(f"\nğŸ’° Calcul du solde pour: {address}")

        for block_index, block in enumerate(self.chain):
            for transaction in block.transactions:
                if transaction.from_addr == address:
                    balance -= transaction.amount
                    print(f"   â– Bloc {block_index}: -{transaction.amount} (â†’ {transaction.to_addr})")
                if transaction.to_addr == address:
                    balance += transaction.amount
                    print(f"   â• Bloc {block_index}: +{transaction.amount} (â† {transaction.from_addr})")

        print(f"   ğŸ¦ Solde final: {balance} unitÃ©s")
        return balance

    def is_chain_valid(self):
        """
        VÃ©rifie l'intÃ©gritÃ© de toute la blockchain
        """
        print("\nğŸ” VÃ‰RIFICATION INTÃ‰GRITÃ‰ BLOCKCHAIN...")

        for i in range(1, len(self.chain)):
            current_block = self.chain[i]
            previous_block = self.chain[i-1]

            # VÃ©rification 1: Le hash stockÃ© correspond-il au hash calculÃ© ?
            if current_block.hash != current_block.create_hash():
                print(f"âŒ Bloc {i} corrompu: hash modifiÃ©")
                return False

            # VÃ©rification 2: Le lien avec le bloc prÃ©cÃ©dent est-il intact ?
            if current_block.previous_hash != previous_block.hash:
                print(f"âŒ Bloc {i} dÃ©connectÃ©: lien brisÃ© avec bloc {i-1}")
                return False

            print(f"âœ… Bloc {i} valide")

        print("ğŸ‰ Toute la blockchain est valide!")
        return True

    def display_blockchain(self):
        """
        Affiche un rÃ©sumÃ© visuel de toute la blockchain
        """
        print("\n" + "="*60)
        print("ğŸ“š BLOCKCHAIN - VUE D'ENSEMBLE")
        print("="*60)

        for i, block in enumerate(self.chain):
            print(f"\nğŸ”— BLOC #{i}")
            print(f"   â° Timestamp: {block.timestamp}")
            print(f"   ğŸ†” Hash: {block.hash[:20]}...")
            print(f"   ğŸ”— Previous Hash: {block.previous_hash[:20]}...")
            print(f"   âš’ï¸  Nonce: {block.nonce}")
            print(f"   ğŸ“‹ Transactions ({len(block.transactions)}):")

            for tx in block.transactions:
                if tx.from_addr is None:
                    print(f"      ğŸ’° RÃ‰COMPENSE: â†’ {tx.to_addr}: {tx.amount}")
                else:
                    print(f"      ğŸ’¸ TRANSFERT: {tx.from_addr} â†’ {tx.to_addr}: {tx.amount}")

# =============================================================================
# Ã‰TAPE 4: DÃ‰MONSTRATION COMPLÃˆTE
# =============================================================================
def demo_blockchain():
    """
    DÃ©monstration complÃ¨te du fonctionnement de la blockchain
    """
    print("ğŸš€ DÃ‰MARRAGE DE LA DÃ‰MONSTRATION BLOCKCHAIN")
    print("="*50)

    # -----------------------------------------------------
    # PHASE 1: INITIALISATION
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 1: INITIALISATION BLOCKCHAIN")
    blockchain = Blockchain()

    # -----------------------------------------------------
    # PHASE 2: CRÃ‰ATION DE TRANSACTIONS
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 2: CRÃ‰ATION DES TRANSACTIONS")

    # Transaction 1: Alice envoie Ã  Bob
    tx1 = Transaction("Alice", "Bob", 50)
    blockchain.create_transaction(tx1)

    # Transaction 2: Bob envoie Ã  Charlie
    tx2 = Transaction("Bob", "Charlie", 25)
    blockchain.create_transaction(tx2)

    # Transaction 3: Charlie envoie Ã  David
    tx3 = Transaction("Charlie", "David", 10)
    blockchain.create_transaction(tx3)

    print(f"ğŸ“Š Total transactions en attente: {len(blockchain.pending_transactions)}")

    # -----------------------------------------------------
    # PHASE 3: MINAGE DU PREMIER BLOC
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 3: MINAGE DU PREMIER BLOC")
    blockchain.mine_pending_transactions("Miner1")

    # -----------------------------------------------------
    # PHASE 4: NOUVELLES TRANSACTIONS
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 4: NOUVELLES TRANSACTIONS")
    tx4 = Transaction("David", "Alice", 5)
    blockchain.create_transaction(tx4)

    # -----------------------------------------------------
    # PHASE 5: MINAGE DU DEUXIÃˆME BLOC
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 5: MINAGE DU DEUXIÃˆME BLOC")
    blockchain.mine_pending_transactions("Miner2")

    # -----------------------------------------------------
    # PHASE 6: VÃ‰RIFICATIONS FINALES
    # -----------------------------------------------------
    print("\nğŸ¯ PHASE 6: VÃ‰RIFICATIONS FINALES")

    # Affichage de la blockchain
    blockchain.display_blockchain()

    # VÃ©rification des soldes
    print("\nğŸ¯ SOLDES DES PARTICIPANTS:")
    addresses = ["Alice", "Bob", "Charlie", "David", "Miner1", "Miner2"]
    for addr in addresses:
        blockchain.get_balance_of_address(addr)

    # Validation de l'intÃ©gritÃ©
    blockchain.is_chain_valid()

    print("\nğŸ‰ DÃ‰MONSTRATION TERMINÃ‰E AVEC SUCCÃˆS!")

# =============================================================================
# EXÃ‰CUTION DE LA DÃ‰MONSTRATION
# =============================================================================
if __name__ == "__main__":
    demo_blockchain()
```
